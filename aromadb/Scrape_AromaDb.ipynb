```python
import re
import requests
import pandas as pd
import bs4
from time import sleep
```

```python
def num2soup(num):
    base_url = "http://bioinfo.cimap.res.in/aromadb/web_compound_detail.php?id=CRMOL"
    url = base_url + str(num)
    response = requests.get(url)
    soup = None    
    soup = bs4.BeautifulSoup(response.text, 'html')
    return soup

def soup2odor(soup):
    name = soup.find('h2').text.split(' ')[0]

    divs1 = soup.find_all('div', {'class': 'span2'})
    divs2 = soup.find_all('div', {'class': 'span5'})

    descriptor = None

    for i, div in enumerate(divs1):
        if 'Fragrance Type' in div.text:
            descriptor = divs2[i].text

    trs = soup.find_all('tr')
    PID = None

    for tr in trs:
        if 'PUBCHEM ID' in tr.text:
            PID = tr.text.split('ID')[1]
    
    return [PID, name, descriptor]
```

```python
odors = []

for n in range(1,1650):
    soup = num2soup(n)
    odors.append(soup2odor(soup))
    print(n, 'done')
    sleep(5)
```

```python
df = pd.DataFrame(odors, columns=['CID', 'Molecule Name', 'Raw Descriptors'])
df
```

```python
df.to_csv('AromaDb_raw.csv')
```
