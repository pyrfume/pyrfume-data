```python
import pandas as pd
import pyrfume
from pyrfume.features import smiles_to_mordred, smiles_to_morgan_sim
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import hashlib
```

```python
molecules = pyrfume.load_data('molecules/molecules.csv')
smiles = molecules['IsomericSMILES']
```

```python
mordred = smiles_to_mordred(smiles.values)
```

```python
mordred.isnull().sum().sum()
```

```python
# Only keep features with <25% missing values
nan_counts = mordred.isnull().sum()
good_features = nan_counts[nan_counts<len(nan_counts)*0.25].index
```

```python
# KNN imputation for remaining missing values
ss = StandardScaler()
knn = KNNImputer(n_neighbors=5)
z = ss.fit_transform(mordred[good_features].values)
%time z = knn.fit_transform(z)
z = ss.inverse_transform(z)
mordred_imputed = pd.DataFrame(z, index=mordred.index, columns=mordred[good_features].columns)
```

```python
mordred_imputed.isnull().sum().sum()
```

```python
mordred_imputed = smiles.to_frame(name='SMILES').join(mordred_imputed, on='SMILES', how='inner').drop('SMILES', axis=1)
```

```python
mordred_imputed.shape
```

```python
# CSV file size is too large for direct storage on the Pyrfume-Data repo
# Instead will save a temporary 'features_actual.csv' that will be hashed and added to pyrfume-data/lfs-cache.csv
# so that pyrfume.load_data() will load the actual data from an external URL
mordred_imputed.to_csv('features_actual.csv')
```

```python
# Create hash of the actual file that will be added to 'features.csv' that will sit on the Pyrfume-Data repo
sha256 = hashlib.sha256()
block_size=256*128

with open('features_actual.csv', 'rb') as f:
    for chunk in iter(lambda: f.read(block_size), b''):
        sha256.update(chunk)

hashed = sha256.hexdigest()     
```

```python
df = pd.DataFrame(['redirect', hashed])
df.to_csv('features.csv', header=False, index=False)
```
